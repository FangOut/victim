```python
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt

x = np.array([4.4,3.9,4.0,4.0,3.5,4.1])
y = np.array([78,74,68,76,73,84])
plt.scatter(x,y)
plt.show
plt.savefig('fig/OldFaithful.png')
```

```python
def S_xy(x,y):
    sum = 0
    for i in range(len(x)):
        c = (x[i] - np.mean(x))*(y[i] - np.mean(y))
        sum = sum + c
   

    return sum
        
```

```python
beta_1_hat = S_xy(x,y) / S_xy(x,x)
beta_0_hat = np.mean(y) - beta_1_hat * np.mean(x)

xs = np.linspace(3.25,4.6,100)
ys = beta_1_hat * xs + beta_0_hat
plt.plot(xs,ys)
plt.scatter(x,y)
plt.show

```
